```python
#Factorial of a number

n = int(input("enter a number: "))
fact = 1
while n > 0:
  fact*= n
  n-= 1
print("Factorial =", fact)
```

```python
# Fibonacci series using while loop
n = int(input("enter a number of terms: "))

a = 0
b = 1
i = 1
while i <= n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("enter a limit: "))
a = 0
b = 1
while a <= limit:
  print(a, end= " ")
  a, b = b, a + b
```

```python
# reverse a number
num = int(input("enter a number: "))
rev = 0
while num > 0:
   digit = num % 10
   rev = rev * 10 + digit
   num //= 10
print("reverse of the number is: ", rev)
```

```python
#Check Palidrome number/string
num = int(input("enter a number: "))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //= 10

if temp == rev:
  print("palidrome")
else:
  print("not palidrome")
```

```python
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("yes")
else:
  print("no")
```

```python
s = input("enter a string: ")
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("yes")
else:
  print("no")
```

```python
#check palindrome number
st = input("enter the string:")
rev = st[::-1]
if st == rev:
  print("palindrome")
else:
  print("not palindrome")
```

```python
#count digits in a number
num = int(input("enter a number: "))
count = 0
while num > 0:
  count += 1
  num //= 10
  print("Number of digits =", count)
```

```python
#search an element in a list
nums =[10, 20, 30, 40, 50]
key = int(input("Enter the element to search: "))
i = 0
found = False
while i < len(nums):
  if nums[i] == key:
    print(f"Element found at index {i}")
    found = True
    break
  i += 1
else:
  print("Element not found in the list")
```
